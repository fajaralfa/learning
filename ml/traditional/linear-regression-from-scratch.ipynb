```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = np.arange(10).reshape(5,2)
plt.plot(dataset[:, 0], dataset[:, 1])
plt.show()
```

```python
class LinearRegression():
    def __init__(self):
        self.b = np.random.randn()
        self.w = np.random.randn()
        self.lr = 0.01

    def fit_gd(self, X, y):
        n = len(X)

        # calculate
        y_preds = self.w * X + self.b

        # measure error
        error = y_preds - y

        # compute gradient
        dw = (error * X).mean()
        db = error.mean()
        
        # update param
        self.w -= self.lr * dw
        self.b -= self.lr * db
        
        pass

    def fit_normal(self, X, y):
        # ensure X is 2D
        X = np.array(X).reshape(-1, 1)
        y = np.array(y)

        # add column of 1s for bias
        X_b = np.hstack([np.ones((X.shape[0], 1)), X])

        # normal equation
        theta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y

        # unpack
        self.b = theta[0]
        self.w = theta[1]

    def predict(self, X):
        return self.w * X + self.b
        pass
```

```python
model = LinearRegression()
```

```python
model.fit_normal(dataset[:, 0], dataset[:, 1])
```
