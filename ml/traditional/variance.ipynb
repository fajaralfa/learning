```python
# Example datasets
X = [1, 2, 3]
Y = [2, 4, 6]

# Step 1: Compute means
mean_X = sum(X) / len(X)
mean_Y = sum(Y) / len(Y)

# Step 2: Compute covariance manually
cov = 0
for i in range(len(X)):
    cov += (X[i] - mean_X) * (Y[i] - mean_Y)

cov /= len(X)  # For population covariance
# If you want sample covariance, divide by len(X) - 1 instead

print("Covariance:", cov)
```
