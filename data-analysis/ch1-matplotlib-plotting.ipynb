```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('./data_jobs/')
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
x = [1,2,3,4]
y = [1,2,3,4]

plt.plot(x,y)
plt.show()
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.month
```

```python
monthly_counts = df.job_posted_month.value_counts()
monthly_counts = monthly_counts.sort_index()
```

```python
plt.plot(monthly_counts)
```

```python
job_counts = df.job_title_short.value_counts()
job_counts.sort_values(ascending=False, inplace=True)
```

```python
plt.barh(job_counts.index, job_counts)
plt.show()
```

```python
plt.bar(job_counts.index, job_counts)
plt.title('Posting by Job Title')
plt.ylabel('Count of Job Postings')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# plot via pandas (still using matplotlib internally)

job_counts.plot(kind='bar')
plt.title('Posting by Job Title')
plt.ylabel('Count of Job Postings')
plt.xlabel('')
plt.xticks(rotation=45, ha='right')
plt.show()
```
