A norm is a function that defines the length of a vector as its distance from the origin.

```python
import numpy as np
```

```python
vect = np.array([1,-2,-3,4])
matrix = np.array([[1,-2,3], [-1,2,3]])
```

## L2 Norm

l2 norm is a euclidean distance of a vector from the origin

```python
np.linalg.norm(vect), (vect * vect).sum() ** .5
```

```python
np.linalg.norm(matrix), (matrix * matrix).sum() ** .5
```

## Unit Vector

unit vector is a vector where the l2 norm of it is 1

```python
unitvect = np.array([0.6, 0.8])
np.linalg.norm(unitvect)
```

## L1 Norm

The L1 norm, also known as the Manhattan or taxicab norm, is defined as the sum of the absolute values of a vector's components.

```python
np.abs(vect).sum()
```

```python
np.abs(matrix).sum()
```

## L2 Squared Norm

l2 squared norm or squared euclidean norm is the sum of the squared values of vector's component
we use it instead of standard l2 norm primarily for computational efficiency:

1. No Square Root: It avoids the expensive sqrt() operation, which is crucial in optimization loops (like Gradient Descent) where it's calculated millions of times.
2. Simpler Derivatives: The derivative of $x^2$ is $2x$, which is linear and easy to compute. The derivative of $\sqrt{x^2}$ involves division by the norm itself, which can complicate calculations and introduce numerical instability near zero.
3. Same Minimizer: Since squaring is a monotonic transformation for non-negative numbers, minimizing the squared L2 norm yields the exact same optimal solution as minimizing the standard L2 norm (just at a different scale), making it ideal for regularization (Ridge Regression) and loss functions (MSE).

```python
(vect * vect).sum()
```

```python
(matrix * matrix).sum()
```

## Max Norm

The max norm (or $L_\infty$ norm) is defined as the maximum absolute value among all elements in a vector or matrix.

```python
np.abs(vect).max()
```

```python
np.abs(matrix).max()
```

## Generalized Lp Norm

The generalized Lp norm (often denoted as the $L_p$ norm) is a unified mathematical framework that defines the "size" or "length" of a vector by raising its components to the power of $p$, summing them, and taking the $p$-th root.

```python
import numpy as np
```

```python
def lp_norm(p, vect):
    if p == np.inf: # max norm
        return np.max(np.abs(vect))
    return (np.abs(vect) ** p).sum() ** (1/p)
```

```python
# l2 norm
lp_norm(2, vect)
```

```python
# l1 norm
lp_norm(1, vect)
```
